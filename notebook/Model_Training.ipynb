```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('cleaned_data.csv')
```

```python
df.head()
```

```python
df_copy=df.drop('day',axis=1)
df_copy=df_copy.drop('month',axis=1)
df_copy=df_copy.drop('year',axis=1)
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(df_copy.corr(),annot=True)
```

```python
df=df_copy
```

```python
df.corr()
```

```python
X=df.drop('FWI',axis=1)
y=df['FWI']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.2)
```

```python
X_train.shape,X_test.shape,y_test.shape,y_train.shape
```

```python
X_train.corr()
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(X_train.corr(),annot=True)
```

```python
# here a probelm of MULTICOLLINEARLITY IS THERE SO TO SOLVE THIS WE HAVE 
```

```python
def correlation(dataset,thresold):
    col_corr=set()
    corr_matrix= dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j])>thresold:
                colname= corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
corr_features=correlation(X_train,0.85)
```

```python
corr_features
```

```python
X_train.drop(corr_features,axis=1,inplace=True)
X_test.drop(corr_features,axis=1,inplace=True)
```

```python
X_train.shape,X_test.shape
```

## Feature Scaling Or Standardization

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train_scaled=scaler.fit_transform(X_train)
```

```python
X_test_scaled= scaler.transform(X_test)
```

```python
plt.subplots(figsize=(15, 5))
plt.subplot(1, 2, 1)
sns.boxplot(data=X_train)
plt.title('X_train Before Scaling')
plt.subplot(1, 2, 2)
sns.boxplot(data=X_train_scaled)
plt.title('X_train After Scaling')
```

# MODEL TRAINING : 

# LINEAR REGRESSION

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
```

```python
regresson = LinearRegression()
```

```python
regresson.fit(X_train_scaled,y_train)
```

```python
y_pred_linear=regresson.predict(X_test_scaled)
mse=mean_absolute_error(y_test,y_pred_linear)
mae=mean_absolute_error(y_test,y_pred_linear)
score=r2_score(y_test,y_pred_linear)
```

```python
mse,mae,score
```

# RIGID REGRESSION

```python
from sklearn.linear_model import Ridge
```

```python
rigid = Ridge()
```

```python
rigid.fit(X_train_scaled,y_train)
```

```python
y_predict_rigid=rigid.predict(X_test_scaled)
mse=mean_squared_error(y_test,y_predict_rigid)
mae=mean_absolute_error(y_test,y_predict_rigid)
score=r2_score(y_test,y_predict_rigid)
```

```python
mse,mae,score
```

# LASSO REGRESSION

```python
from sklearn.linear_model import Lasso
```

```python
lasso = Lasso()
```

```python
lasso.fit(X_train_scaled,y_train)
```

```python
y_pred_lasso=lasso.predict(X_test_scaled)
```

```python
mae= mean_absolute_error(y_test,y_pred_lasso)
mse = mean_squared_error(y_test,y_pred_lasso)
score= r2_score(y_test,y_pred_lasso)
```

```python
mse,mae, score
```
