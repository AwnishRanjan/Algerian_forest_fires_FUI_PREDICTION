```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df=pd.read_csv('Algerian_forest_fires_dataset.csv',header=1)
df.head(2)
```

```python
df.columns
```

```python
df.info()
```

```python
# Data cleaning 
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.loc[:122,'region']=0
df.loc[122: ,'region']=1
```

```python
df.head()
```

```python
df['region']=df['region'].astype(int)
```

```python
df.head()
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.dropna().reset_index(drop=True)
```

```python
df.info()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.columns=df.columns.str.strip()
```

```python
df.columns
```

```python
df.head(1)
```

```python
df[['day','month','year','Temperature','RH','Ws']]=df[['day','month','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info()
```

```python
obj = [i for i in  df.columns if df[i].dtypes=='O']
```

```python
obj
```

```python
for i in obj:
    if i !='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df['Classes'].value_counts()
```

```python
df['Classes']=np.where(df['Classes'].str.contains('not fire'),0,1)
```

```python
df.info()
```

```python
#EDA
```

```python
df_copy=df.drop(['day','month','year'],axis=1)
df_copy.head(1)
```

```python
df_copy.hist(bins=50,figsize=(20,15))
```

```python
percentage= df_copy['Classes'].value_counts(normalize=True)*100
percentage
```

```python
label=['fire','not fire']
plt.figure(figsize=(12,7))
plt.pie(percentage,labels=label,autopct='%1.1f%%')
plt.title("Pie Chart of Classes")
plt.show()
```

```python
sns.countplot(data = df, x='month',hue='Classes')
plt.ylabel('Number of Fires',weight='bold')
plt.xlabel('Months',weight='bold')
plt.title("Fire Analysis of All Regions",weight='bold')
```

Its observed that August and September had the most number of forest fires for both regions. And from the above plot of months, we can understand few things

Most of the fires happened in August and very high Fires happened in only 3 months - June, July and August.
